```python
import pandas as pd
import numpy as np
from math import fabs, floor
from random import normalvariate, randint
```

```python
data = pd.DataFrame()
```

```python
y = [randint(0, 1) for i in range(Length)]
data['Итог'] = y
```

```python
Length = 10000
```

```python
salary = []

for i in range((Length)):
    if y[i] == 1:
        salary.append(fabs(round(normalvariate(100, 50))) * 1000)
    else:
        salary.append(fabs(round(normalvariate(30, 10))) * 1000)
        
data['Зарплата'] = salary
```

```python
age = []

for i in range(Length):
    if y[i] == 1:
        age.append(fabs(round(normalvariate(45, 10))) )
    else:
        age.append(fabs(round(normalvariate(30, 10))))
        
for i in range(Length):
    if age[i] < 18:
           age[i] += randint(18, 30)
    if age[i] > 60 and y == 1 and summ[i] > 100000:
        age[i] -= randint(10, 30)
        
        
data['Возраст'] = age
```

```python
summ = []

for i in range(Length):
    if salary[i] > 100000:
        summ.append(fabs(round(normalvariate(500, 100))) * 1000)
    elif salary[i] < 100000 and y[i] == 0:
        summ.append(fabs(round(normalvariate(500, 100))) * 1000)
    else:
        summ.append(fabs(round(normalvariate(200, 50))) * 1000)
    

        
for i in range(Length):
    if summ[i] < 5000:
           summ[i] += randint(0, 10) * 1000
        
        
data['Сумма'] = summ
```

```python
time = []
time_array = [12, 18, 24, 30, 36, 42, 48, 54, 60]

for i in range(Length):
    if summ[i] > 200000:
        time.append(time_array[randint(2, len(time_array) - 1)])
    else:
        time.append(time_array[randint(0, len(time_array)-2)])

        
for i in range(Length):
    if time[i] < 12:
           time[i] += randint(0, 10)
        
        
data['Срок'] = time
```

```python
exp = []
exp_array = [6 * i for i in range(50)]

for i in range(Length):
    if age[i] > 60:
        exp.append(exp_array[randint(30, len(exp_array) - 1)])
    elif age[i] <= 60 and age[i] > 40:
        exp.append(exp_array[randint(5, 45)])
    elif age[i] <= 40 and age[i] > 25:
        exp.append(exp_array[randint(5, 16)])
    else:
        exp.append(exp_array[randint(1, 10)])

        
for i in range(Length):
    if age[i] < 20 and exp[i] > 4:
           exp[i] -= exp_array[randint(1, 3)]
        
        
data['Опыт'] = exp
```

```python
history = []

for i in range(Length):
    if y[i] == 1:
        history.append(fabs(round(normalvariate(3, 1))) )
    else:
        history.append(fabs(round(normalvariate(0, 1))))
        
        
data['Кредитная история'] = history
```

```python
current = []

for i in range(Length):
    if y[i] == 1:
        current.append(fabs(floor(normalvariate(0, 0.01))))
    else:
        current.append(fabs(floor(normalvariate(0, 2))))
        
        
data['Текущие кредиты'] = current
```

```python
overdue = []

for i in range(Length):
    if y[i] == 1:
        overdue.append(fabs(round(normalvariate(0, 0))) )
    else:
        overdue.append(fabs(round(normalvariate(0, 1))))
        
        
data['Непогашенные кредиты'] = overdue
```
